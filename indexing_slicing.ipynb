# **Indexing & Slicing**

```python
#Access the 5th element of a 1D array
import numpy as np

array_1d = np.array([3,5,1,2,9,7,4,6])

print(array_1d[4])
```

```python
#Slice first 5 elements.
print(array_1d[:5])
```

```python
#Extract last 3 elements
print(array_1d[-3:])
```

```python
#Reverse a NumPy array
print(array_1d[::-1])
```

```python
#Access element at row 2, column 3 from a 2D array

array_2d = np.array([ [2,4,5,7],
                      [5,6,7,9],
                      [4,7,7,6] ])
array_2d.shape

print(array_2d[2,3])
```
