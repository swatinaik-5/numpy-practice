# **Create Arrays**

```python
#Create a 1D NumPy array from a Python list.
import numpy as np
array = np.array([2,5,6,7,8])
array
```

```python
#Create a NumPy array of zeros of size 10.
zeros_arr = np.zeros(10, dtype=int)
zeros_arr
```

```python
#Create a NumPy array of ones of shape (3, 4)
ones_arr = np.ones([3,4], dtype=int)
ones_arr
```

```python
#Create an array with values from 5 to 50 (step = 5).
array = np.arange(5,51,5)
array
```

```python
#Create an array of 20 random integers between 1 and 100.
array = np.random.randint(1,100,20)
array
```

# **Array Properties**

```python
#Find shape, size, ndim, dtype of an array
array = np.array([2,8,9,3,5])
print("Shape of an array: ", array.shape)
print("Size of an array: ", array.size)
print("Dimension of an array: ", array.ndim)
print("Data type of an array: ", array.dtype)
```

```python
#Convert an integer array to float type.

array = array.astype(float)
array
```

```python
#Check memory size of an array
array.nbytes
```
